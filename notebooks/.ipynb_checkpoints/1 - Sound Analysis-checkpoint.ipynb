# 1 - Sound Analysis

This notebook displays the analysis GUI, as right now every file must be pre-analyzed before being able to use it in the morphing plugin, this tool is the one that generate the ".had" files, necessary for the plugin to correctly extract information from the sound file. The steps to generate the ".had" file are the following:

1. Try to guess the parameters
2. Run the analysis and hear the results
3. If you are not happy with them, tweak the parameters and run the analysis again
4. Once you think the result of the synthesis is good, click the "Generate .had file" button and we are done here!
